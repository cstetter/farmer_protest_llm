```python
import pandas as pd

df_protests = pd.read_csv('../data/2019_01_01_2024_09_10_acled_europe_raw.csv', sep=',')

# remomve trailing white space from column names
df_protests.columns = df_protests.columns.str.strip()

# Extract month
# Assuming 'event_date' is the column with the date format like '30 August 2024', convert date column to datetime format
df_protests['event_date'] = pd.to_datetime(df_protests['event_date'], format='%d %B %Y')

# Create new column 'month'
df_protests['month'] = df_protests['event_date'].dt.month

df_protests
```

```python
import geopandas as gpd

grouped_data = df_protests.groupby(['year', 'month', 'longitude', 'latitude', 'country']).size().reset_index(name='count')

gdf = gpd.GeoDataFrame(
    grouped_data, geometry=gpd.points_from_xy(grouped_data.longitude, grouped_data.latitude), crs="EPSG:4326"
)

# Step 1: Calculate the sum of points for each year/month combination
point_sums = gdf.groupby(['year', 'month']).count()['count'].reset_index()

# Step 2: Merge the sums back with the original GeoDataFrame
gdf = gdf.merge(point_sums, on=['year', 'month'], suffixes=('', '_sum'))


gdf['count_sum'] = chr(8721) + "=" + gdf['count_sum'].astype(str)



# Define month names
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Map numerical months to names
gdf['month'] = gdf['month'].map(month_names)

# Ensure that 'month_name' is a categorical variable with the correct order
gdf['month'] = pd.Categorical(gdf['month'], categories=month_names.values(), ordered=True)

gdf
```

```python
from shapely.geometry import Polygon

# Load europe
zipfile = "../data/ne_110m_admin_0_countries.zip"
world = gpd.read_file(zipfile)

europe = world[world.CONTINENT == 'Europe'].iloc[1:,] #Exclude Russia


# Make polygon from bbox coordinates https://stackoverflow.com/a/68741143/18253502
def make_bbox(long0, lat0, long1, lat1):
    return Polygon([[long0, lat0],
                    [long1,lat0],
                    [long1,lat1],
                    [long0, lat1]])

# Coords covering Europe & Russia made with http://bboxfinder.com
bbox = make_bbox(-36.386719,29.228890,60.292969,74.543330)

# STEP 2 #
# Convert to gdf
bbox_gdf = gpd.GeoDataFrame(index=[0], crs='epsg:4326', geometry = [bbox])

# Use bbox as clipping border for Europe
europe = europe.overlay(bbox_gdf, how="intersection")
```

# Proposer

Split df for proposer to make managable for LLM

```python
# Set seed for reproducibility
seed = 42

# Shuffle the rows randomly
df_shuffled = df_main.sample(frac=1, random_state=seed).reset_index(drop=True)

# Split the data into 3 approximately equal parts
split_size = len(df_shuffled) // 3

# First, second, and third random subsamples
subsample_1 = df_shuffled.iloc[:split_size]
subsample_2 = df_shuffled.iloc[split_size:2*split_size]
subsample_3 = df_shuffled.iloc[2*split_size:]

description_subsample_1 = ';'.join(subsample_1['description'].fillna(''))
description_subsample_2 = ';'.join(subsample_2['description'].fillna(''))
description_subsample_3 = ';'.join(subsample_3['description'].fillna(''))

description_list = [description_subsample_1, description_subsample_2, description_subsample_3]
```

# Make prompts

## Prompt strategies

* **bothinst**: Split instructions with role as a system msg and task as a user msg
* **zero-CoT**: Ask the model to reason step-by-step before providing its answer.
* **strict**: Ask the model to answer by strictly following a given template.
* **info**: Provide additional information to address common reasoning failures.
* name
* pos



```python
system_prompt = '''
You are Bob, an intelligent expert at categorizing textual information on farmer protests in Europe between 2023 and 2024.
You are tasked with discerning key issues, a.k.a. categories, from textual descriptions and reasons of farmer protests.
Make the categories diverse and distinguishable across, and coherent within.
Output: You provide categories and a concise description for each category in table format only.
'''

# user1_prompt = '''I will give you some text and you will analyse it, and refine it step-by-step by at least three steps. Got it?'''
user1_prompt = '''I will give you some text and you will analyse it. Got it?'''

assistant1_prompt = "Yes, I understand. I am Bob, and I will analyse the given task as best as I can."

def make_user_prompt(text):
    prompt = '''Great! Let's begin then :). Analyse the following text: \n\n %s \n
    Answer: Let's reach a comprehensive categorization''' %(text)

    return prompt
```

```python
import openai
import os
from dotenv import load_dotenv

load_dotenv()
openai.api_key = os.getenv('openai_api_key') # Add your own API key
```

```python
responses = []

for n in range(5): # Make 5 runs each subsample
  for i in description_list:
    responses.append(openai.chat.completions.create(model="gpt-4o",
                                          messages=[{"role": "system", "content": system_prompt},
                                                    {"role": "user", "content": user1_prompt},
                                                    {"role": "assistant", "content": assistant1_prompt},
                                                    {"role": "user", "content": make_user_prompt(i)}],
                                                    temperature=1,
                                                    seed=42
                                                    )
      )
```
